```python
# Dependencies and Setup
import pandas as pd
```

```python
# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"
```

```python
# Read School and Student Data File and store into Pandas DataFrames
school_data_df = pd.read_csv(school_data_to_load)
student_data_df = pd.read_csv(student_data_to_load)
```

```python
# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data_df, school_data_df, how="left", on=["school_name", "school_name"])
```

```python
#1a) Create a df listing all of the unique schools
school = ["school_name"]
school
school_data_df.loc[:, school]
```

```python
#1b) Count each school in the newly created dataframe
total_schools = school_data_df["school_name"].count()
total_schools
```

```python
#2a) Create a df listing all of the unique student names
students = ["student_name"]
students
student_data_df.loc[:, students]
```

```python
#2b) Total number of students
total_students = school_data_df["size"].sum()
total_students
```

```python
#3a) Create a df for the budgets of each school
budgets = ["budget"]
budgets
school_data_df.loc[:, budgets]
```

```python
#3b) Calculate total school budget
total_budget = school_data_df["budget"].sum()
total_budget
```

```python
#4a) Create a df for all of the math scores
math_scores = ["math_score"]
math_scores
student_data_df.loc[:, math_scores]
```

```python
#4b) Calculate the average math score
math_avg = student_data_df["math_score"].mean()
math_avg
```

```python
#5a) Created a df for all of the reading scores
reading_scores = ["reading_score"]
reading_scores
student_data_df.loc[:, reading_scores]
```

```python
#5b) Average readings score
reading_avg = student_data_df["reading_score"].mean()
reading_avg
```

```python
#6a) Create a dataframe of math scores >= 70 (True/False)
math_pass = ["math_score"]
math_pass
student_data_df.loc[:, math_scores] >=70
```

```python
#6b) Passing Math (The percentage of students that passed math.)
math_pass_tot = student_data_df["math_score"].loc[student_data_df["math_score"] >=70].count() 
math_pass_rate = (math_pass_tot/total_students)*100
math_pass_rate
```

```python
#7a) Create a dataframe of math scores >= 70 (True/False)
math_pass = ["math_score"]
math_pass
student_data_df.loc[:, math_scores] >=70
```

```python
#7b) Passing Reading (The percentage of students that passed reading.)
read_pass_tot = student_data_df["reading_score"].loc[student_data_df["reading_score"] >=70].count()
read_pass_rate = (read_pass_tot/total_students)*100
read_pass_rate
```

```python
#8a) Create a dataframe of reading scores >= 70
math_read_pass_df = ["math_score","reading_score"]
math_read_pass_df
student_data_df.loc[(student_data_df["reading_score"] >=70) & (student_data_df["math_score"] >=70)]
```

```python
#8b) Overall Passing (The percentage of students that passed math and reading.)
#Split into two -- math pass rate and read pass rate
math_read_pass_rate = student_data_df.loc[(student_data_df["reading_score"] >=70) & (student_data_df["math_score"] >=70),:].mean()
math_read_pass_rate
```

```python
# create dataframe to hold results
summary_table = pd.DataFrame(
    {"Total Schools":[total_schools],
     "Total Students":[total_students],
     "Total Budget":[total_budget],
     "Average Math Score":[math_avg],
     "Average Reading Score":[reading_avg],
     "% Passing Math":[math_pass_rate],
     "% Passing Reading":[read_pass_rate],
     "% Overall Passing": [math_read_pass_rate]})
summary_table 
```
